```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Load the Excel file
df = pd.read_excel('/content/PMEA_37_12_2181_OnlineAppendix.xlsx', sheet_name='SUAHSDB')
df = df[['Record name', 'Diagnosis', 'Type (-1=normal 1=abnormal)']]

# Load the audio files
sample_rate1, audio_data1 = wavfile.read('/content/f0096.wav')
print('The sampling rate of this signal1: ', sample_rate1)

sample_rate2, audio_data2 = wavfile.read('/content/f0015.wav')
print('The sampling rate of this signal2: ', sample_rate2)

# Create time arrays using np.arange
time1 = np.arange(0, len(audio_data1) / sample_rate1, 1 / sample_rate1)
time2 = np.arange(0, len(audio_data2) / sample_rate2, 1 / sample_rate2)

plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
plt.plot(time1, audio_data1)
plt.title('Audio Waveform [Normal]')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(time2, audio_data2)
plt.title('Audio Waveform [Abnormal]')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
